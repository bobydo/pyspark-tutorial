```python
# Set the PySpark environment variables
import os
import sys
from dotenv import load_dotenv
import pyspark

load_dotenv()  # loads .env into os.environ

# Ensure Hadoop bin is on PATH (needed on Windows)
os.environ["PATH"] = os.environ["HADOOP_BIN"] + ";" + os.environ["PATH"]

print(sys.executable)
print(os.environ["HADOOP_HOME"])
print(os.environ["JAVA_HOME"])
print(sys.executable)
print(pyspark.__version__)
```

```python
# pyright: reportMissingImports=false
from pyspark.sql import SparkSession
# Use local[1] instead of local[*] on Windows.
# local[*] spawns multiple Python workers, which can crash due to
# Windows process, file-locking, and IPC limitations.

# Disable Arrow and Python worker reuse for stability on native Windows.
# These features improve performance but can crash Python workers due to
# Windows process, file-locking, and IPC limitations.

spark = (
    SparkSession.builder
    .master("local[1]")   # ðŸ‘ˆ IMPORTANT: NOT local[*]
    .appName("test")
    .config("spark.sql.execution.arrow.pyspark.enabled", "false")
    .config("spark.python.worker.reuse", "false")
    .getOrCreate()
)


data = [("Alice", 25), ("Bob", 30), ("Charlie", 35)]
df = spark.createDataFrame(data, ["Name", "Age"])
df.show(vertical=False)
```
