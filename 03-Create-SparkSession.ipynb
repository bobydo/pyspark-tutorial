## Create SparkSession in Apache Spark

```python
from pyspark.sql import SparkSession

# Create a SparkSession
# Each executor gets 2 GB RAM

# Intel i5-8600.Physical cores: 6 
# After a shuffle, Spark will create 4 partitions.
# Each partition = one task.
# How this relates to cores
# One task runs on one core
# Max parallel tasks = number of available cores

spark = SparkSession.builder \
    .appName("MySparkApplication") \
    .config("spark.executor.memory", "2g") \
    .config("spark.sql.shuffle.partitions", "4") \
    .getOrCreate()
```

```python
# Perform operations using the SparkSession
spark
```

```python
# Shut down the current active SparkSession
spark.stop()
```
